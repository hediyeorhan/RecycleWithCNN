```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```python
# kullanılacak veri artırma tekniklerini tanımla
datagen = ImageDataGenerator(rotation_range=40,
                             width_shift_range=0.2,
                             height_shift_range=0.2,
                             shear_range=0.2,
                             zoom_range=0.2,
                             horizontal_flip=True,
                             vertical_flip=True,
                             zca_whitening=True,
                             fill_mode='nearest',
                             featurewise_center=True, samplewise_center=True,
                              featurewise_std_normalization=True, 
                             samplewise_std_normalization=True,
                             brightness_range=[0.5,2.5]) 
```

```python
import os, sys, glob

folder = "/content/drive/MyDrive/DataV4/test/plastic"

for filename in glob.iglob(os.path.join(folder, "*.jpg")):
       infilename = os.path.join(folder,filename)
       print(infilename)
       img = load_img(infilename)
       x = img_to_array(img)
       x = x.reshape((1,) + x.shape)
       i = 0

       for batch in datagen.flow(x, batch_size=1,
                          save_to_dir='ArtirilmisVeri/test/plastic', 
                          save_format='jpg'):
         
         
         i += 1
         if i > 20:
           break 

```
